#Aim
Analyze and record power usage

```python
import pandas as pd

# Load the our csv
df = pd.read_csv('/content/sample_data/watt_usage.csv')

# Convert Date column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Sort data by timestamp to ensure correct calculation of consumption
df = df.sort_values(by='Date').reset_index(drop=True)

# Display the first 5 rows of the DataFrame and its information
print(df.head())
df.info()
```

```python
# Convert 'Units' column to numeric
df['Units'] = df['Units'].astype('float')

# Display the first 5 rows of the DataFrame and its information to confirm the change
print(df.head())
df.info()
```

## Analyze Usage


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot of Units against Date
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Date', y='Units', data=df)
plt.title('Units Readings Over Time')
plt.xlabel('Date')
plt.ylabel('Units')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
